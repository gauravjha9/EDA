## EDA on Flight Price

```python
# import some important Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Train data set

train_df = pd.read_excel('data/Data_Train.xlsx')
train_df.head()
```

```python
# Load Test data set

test_df = pd.read_excel('data/Data_Test.xlsx')
test_df.head()
```

```python
# merge both dataset (Train and Test)

flight_df = train_df.append(test_df)
flight_df.head()
```

```python
flight_df.info()
```

```python
# create different column for Date.

flight_df['Date'] = flight_df['Date_of_Journey'].str.split('/').str[0]
flight_df['Month'] = flight_df['Date_of_Journey'].str.split('/').str[1]
flight_df['Year'] = flight_df['Date_of_Journey'].str.split('/').str[2]
```

```python
# another technique for the same.

# flight_df['Date'] = flight_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
# flight_df['Month'] = flight_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
# flight_df['Year'] = flight_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
flight_df.head(2)
```

```python
# Convert Date, Month, Year column's datatype to int

flight_df['Date']=flight_df['Date'].astype(int)
flight_df['Month']=flight_df['Month'].astype(int)
flight_df['Year']=flight_df['Year'].astype(int)
```

```python
# Drop Date_of_Journey column

flight_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
flight_df.head(2)
```

```python
# Remove Date from Arrival_Time

flight_df['Arrival_Time']=flight_df['Arrival_Time'].str.split(' ').str[0]
```

```python
# Create Columns Arrival_hour, Arrival_min for Arrival_Time

flight_df['Arrival_hour'] = flight_df['Arrival_Time'].str.split(':').str[0]
flight_df['Arrival_min'] = flight_df['Arrival_Time'].str.split(':').str[1]
```

```python
flight_df.head(2)
```

```python
# Convert Arrival_hour, Arrival_min into int data type

flight_df['Arrival_hour'] = flight_df['Arrival_hour'].astype(int)
flight_df['Arrival_min'] = flight_df['Arrival_min'].astype(int)
```

```python
# Drop Arrival_Time

flight_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
flight_df.head(1)
```

```python
# Create Columns Dep_hour, Dep_min for Dep_Time

flight_df['Dep_hour'] = flight_df['Dep_Time'].str.split(':').str[0]
flight_df['Dep_min'] = flight_df['Dep_Time'].str.split(':').str[1]
```

```python
# Convert Dep_hour, Dep_min to int data type

flight_df['Dep_hour'] = flight_df['Dep_hour'].astype(int)
flight_df['Dep_min'] = flight_df['Dep_min'].astype(int)
```

```python
flight_df.head(2)
```

```python
# Drop Dep_Time column

flight_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
flight_df['Total_Stops'].unique()
```

```python
# Replace Total_Stops string values to number.

flight_df['Total_Stops'] = flight_df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, 'nan':1})
```

```python
flight_df.head(2)
```

```python
# Drop Route Column

flight_df.drop('Route', axis=1, inplace=True)
```

```python
# Find some value Duration column that is 5m

flight_df[flight_df['Duration']== '5m']
```

```python
# Drop rows which have 5m Duration value

flight_df.drop([6474, 2660], inplace=True)
```

```python
# split Duration's hour value to Duration_min.

duration_hour_to_min = flight_df['Duration'].str.split(' ').str[0].str.split('h').str[0].astype(int) * 60
duration_hour_to_min
```

```python
# Duration min

duration_min = flight_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
duration_min
```

```python
# Replace nan value to 0 min

duration_min.replace({np.nan: 0}, inplace=True)
```

```python
duration_min
```

```python
# convert duration min to int datatype and store in a variable

duration_min = duration_min.astype(int)
```

```python
# create Duration_minutes column and add duration_hour_to_min with duration_min

flight_df['Duration_minutes'] = duration_hour_to_min + duration_min
```

```python
flight_df.head(4)
```

```python
# Drop Duration column

flight_df.drop('Duration', axis=1, inplace=True)
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
# Label encoding for the remaining categorical column

flight_df['Airline'] = labelencoder.fit_transform(flight_df['Airline'])
flight_df['Source'] = labelencoder.fit_transform(flight_df['Source'])
flight_df['Destination'] = labelencoder.fit_transform(flight_df['Destination'])
flight_df['Additional_Info'] = labelencoder.fit_transform(flight_df['Additional_Info'])
```

```python
flight_df.head(2)
```

```python
# one hot encoding using get_dummies

pd.get_dummies(flight_df, columns=['Airline', 'Source', 'Destination', 'Additional_Info'], drop_first=True)
```
